```python
def make_frame_list(lines,enviroment,label):

    lines = open(lines, "r")
    whole_file = lines.readlines()#makes a list of strings, each element is a line in txt file
    #desired_lines = whole_file[1::6]#makes a new array that made up of every 6th line from the 2nd line

    print(list(whole_file[7]))
    print(list(whole_file[12]))
    print(list(whole_file[14]))
    print(list(whole_file[15]))
    
    
    object_no = 7
    radial = 12
    snr =  14
    noise = 15
    
    res = []
    while object_no < len(whole_file):

        #string current looks like '5,38\n'
        
       
        sub_object = whole_file[object_no].strip("\n")#removes newline charcter
        sub_object = sub_object.split(":")
        


        #string would now be '5,38'
        sub_radial_distance = whole_file[radial].strip("\n")
        sub_radial_distance = sub_radial_distance.split(":")


        sub_snr = whole_file[snr].strip("\n")
        sub_snr = sub_snr.split(":")

        sub_noise = whole_file[noise].strip("\n")
        sub_noise = sub_noise.split(":")


        #turns string into list ['5','38']


        #turns each element into int
        # sub[0] = int(sub[0])
        # sub[1] = int(sub[1])

        object_no+=15
        radial+=15
        snr+= 15
        noise += 15


        sub_object = list(sub_object)[-1].split(",")
        sub_radial_distance = list(sub_radial_distance)[-1].split(",")
        sub_snr = list(sub_snr)[-1].split(",")
        sub_noise = list(sub_noise)[-1].split(",")


        sub_object = int(sub_object[0])
        sub_radial_distance = [float(x) for x in sub_radial_distance]
        sub_snr = [float(x) for x in sub_snr]
        sub_noise = [float(x) for x in sub_noise]

       
      


        #adds new list into main list
        res.append([sub_object,sub_radial_distance,sub_snr,sub_noise,enviroment,label])


    return res

make_frame_list('dark_dime1.txt','dark','dime')
```
